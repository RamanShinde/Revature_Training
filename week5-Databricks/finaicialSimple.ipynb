```python
import pandas as pd
pdf=pd.read_excel("/Volumes/workspace/default/myvolume/Financial Sample.xlsx")
df=spark.createDataFrame(pdf)
display(df)
```

```python
for col_name in df.columns:
    df=df.withColumnRenamed(col_name,col_name.replace(" ","_"))
```

```python
from pyspark.sql.functions import col
nums_cos=[
    "Manufacturing_Price",
    "Sale_Price",
    "Gross_Sales",
    "Discounts",
    "_Sales",
    "COGS",
    "Profit"
]

for c in nums_cos:
    if c in df.columns:
         df=df.withColumn(c,col(c).cast("double"))

df=df.withColumn("Discount_Band",col("Discount_Band").cast("string"))
```

```python
df_silver=df.dropna()
display(df_silver)
```

```python
from pyspark.sql import functions as f

df_gold=df_silver.groupBy("Country").agg(
    f.sum("_Sales").alias("Total_Sales"),
    f.sum("Profit").alias("Total_Profit"),
    f.sum("COGS").alias("Total_COGS"),
    f.avg("Discounts").alias("discount")
)
display(df_gold)
```
